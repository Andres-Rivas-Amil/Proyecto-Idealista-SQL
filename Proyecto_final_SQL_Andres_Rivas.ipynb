## Primera parte

### 1. ¿Cuáles son las propiedades que tienen al menos 2 baños, un ascensor y una plaza de aparcamiento? Selecciona el ASSETID, BATHNUMBER, HASLIFT, HASPARKINGSPACE y PRICE

```python
sSQL = """

SELECT ASSETID, BATHNUMBER AS Baños, HASLIFT AS Ascensores, HASPARKINGSPACE AS Aparcamiento, price AS precio
FROM idealista
WHERE bathnumber >= 2
AND HASLIFT = 1
AND hasparkingspace = 1

"""
pd.read_sql(sSQL, con=conn)
```

### 2. ¿Cuál es el precio promedio de las propiedades con terraza?

```python
sSQL = """

SELECT AVG(price) AS Precio_medio_propiedades_con_terraza
FROM idealista
WHERE hasterrace =1

"""
pd.read_sql(sSQL, con=conn)
```

### 3. ¿Cuántas propiedades hay según el número de habitaciones? ¿Cuál es el precio medio de los apartamentos según el número de habitaciones? Ordénalo por número de habitaciones de menor a mayor.

```python
sSQL = """

SELECT roomnumber as Numero_habitaciones, COUNT(*) AS total_propiedades, AVG(price) AS precio_medio
FROM idealista
GROUP BY roomnumber
ORDER BY roomnumber ASC

"""
pd.read_sql(sSQL, con=conn)
```

### 4. ¿Cuáles son las propiedades cuyo precio es mayor que el precio promedio de todas las propiedades en el dataset? Muestra el ASSETID, PRICE y CONSTRUCTEDAREA

```python
sSQL = """

SELECT assetid, price AS precio, constructedarea AS Tamaño_propiedad
FROM idealista
WHERE price > (SELECT AVG(price)FROM idealista)

"""
pd.read_sql(sSQL, con=conn)
```

### 5. ¿Cuántas propiedades tienen más de 3 habitaciones y están situadas en la última planta del edificio?

```python
sSQL = """

SELECT COUNT(*) ASSETID
FROM idealista
WHERE roomnumber > 3 AND isintopfloor = 1

"""
pd.read_sql(sSQL, con=conn)
```

### 6. ¿Cuál es la propiedad más cara registrada en el dataset, es decir, su ASSETID?

```python
sSQL = """

SELECT assetid, price AS Precio_propiedad_mas_cara
FROM idealista
ORDER BY price DESC
LIMIT 1

"""
pd.read_sql(sSQL, con=conn)
```

### 7. ¿Cuál es la distancia promedio al centro de la ciudad para las propiedades construidas después del año 2000?

```python
sSQL = """

SELECT AVG(DISTANCE_TO_CITY_CENTER) AS Distancia_promedio_a_centro_ciudad
FROM idealista
WHERE CADCONSTRUCTIONYEAR > 2000

"""
pd.read_sql(sSQL, con=conn)
```

### 8. ¿Cuántas propiedades tienen piscina, jardín y plaza de aparcamiento?

```python
sSQL = """

SELECT COUNT(*) AS Total_propiedades_con_piscina_jardin_y_plaza_aparcamiento
FROM idealista
WHERE HASSWIMMINGPOOL = 1
AND HASGARDEN = 1
AND HASPARKINGSPACE = 1

"""
pd.read_sql(sSQL, con=conn)
```

### 9. ¿Qué porcentaje de propiedades de segunda mano están en buen estado y tienen más de 2 baños?

```python
sSQL = """

SELECT COUNT(*) * 100.0 / (SELECT COUNT(*) FROM idealista WHERE bathnumber > 2) AS porcentaje
FROM idealista
WHERE bathnumber > 2 AND builttypeid_3 = 1

"""
pd.read_sql(sSQL, con=conn)
```

### 10. ¿Hay alguna tendencia en el valor medio de las propiedades en los distintos meses del año 2018? ¿Y en el valor máximo o mínimo?

```python
sSQL = """
SELECT  
    MONTH(STR_TO_DATE(CONCAT(PERIOD, '01'), '%Y%m%d')) AS mes, 
    AVG(PRICE) AS precio_medio,
    MAX(PRICE) AS precio_maximo, 
    MIN(PRICE) AS precio_minimo
FROM idealista
WHERE YEAR(STR_TO_DATE(CONCAT(PERIOD, '01'), '%Y%m%d')) = 2018
GROUP BY mes
ORDER BY mes
"""

pd.read_sql(sSQL, con=conn)
```

### 14. ¿Cuál es el precio promedio por metro cuadrado para propiedades con orientación sur, este y oeste?

```python
sSQL = """

SELECT 
    'Sur' AS ORIENTATION, 
    AVG(PRICE / CONSTRUCTEDAREA) AS precio_promedio_m2
FROM idealista
WHERE HASSOUTHORIENTATION = 1
UNION ALL
SELECT 
    'Este' AS ORIENTATION, 
    AVG(PRICE / CONSTRUCTEDAREA) AS precio_promedio_m2
FROM idealista
WHERE HASEASTORIENTATION = 1
UNION ALL
SELECT 
    'Oeste' AS ORIENTATION, 
    AVG(PRICE / CONSTRUCTEDAREA) AS precio_promedio_m2
FROM idealista
WHERE HASWESTORIENTATION = 1

"""
pd.read_sql(sSQL, con=conn)
```

### 15. ¿Cuál es la diferencia de los precios promedios entre propiedades que incluyen la plaza de aparcamiento y las que no?

```python
sSQL = """

SELECT
    AVG(CASE WHEN HASPARKINGSPACE = 1 then price else NULL END) AS Precio_promedio_con_aparcamiento,
    AVG(CASE WHEN HASPARKINGSPACE = 0 then price else NULL END) AS Precio_promedio_sin_aparcamiento,
    AVG(CASE WHEN HASPARKINGSPACE = 1 then price else NULL END) - AVG(CASE WHEN HASPARKINGSPACE = 0 then price else NULL END) AS Diferencia_precio_promedio
FROM idealista

"""
pd.read_sql(sSQL, con=conn)
```

## Segunda parte

### 1. ¿Qué propiedades gestionadas por agencias tienen piscina y plaza de aparcamiento?

```python
sSQL = """

SELECT 
    i.ASSETID, 
    i.AGENCY_ID, 
    a.AGENCY_NAME AS Nombre_Agencia,
    i.HASSWIMMINGPOOL as Piscina,  
    i.HASPARKINGSPACE AS Plaza_aparcamiento
FROM idealista i
INNER JOIN agencias a 
    ON i.AGENCY_ID = a.AGENCY_ID
WHERE 
    i.AGENCY_ID IS NOT NULL
    AND i.HASSWIMMINGPOOL = 1
    AND i.HASPARKINGSPACE = 1;
"""


pd.read_sql(sSQL, con=conn)
```

### 2. ¿Cuáles son todas las propiedades gestionadas por una agencia con puntuación mayor a 4.5? Muestra el ASSETID, AGENCY_NAME, PRICE y RATING.

```python
sSQL = """

SELECT 
    i.ASSETID,
    a.AGENCY_NAME,
    i.PRICE,
    a.RATING
FROM idealista AS i
INNER JOIN agencias AS a
ON  i.AGENCY_ID = a.AGENCY_ID
WHERE  a.RATING > 4.5

"""
pd.read_sql(sSQL, con=conn)
```

### 3. ¿Qué inmobiliarias no tienen propiedades listadas en el dataset?

```python
sSQL = """

SELECT AGENCY_ID
FROM idealista
WHERE AGENCY_ID IS NULL OR AGENCY_ID NOT IN (1, 2, 3, 4, 5)

"""
pd.read_sql(sSQL, con=conn)
```

### 4. ¿Cuál es la inmobiliaria con el precio promedio más alto en propiedades listadas?

```python
sSQL = """

SELECT a.AGENCY_NAME as Nombre_agencia, COALESCE(AVG(i.PRICE), 0) AS Precio_promedio
FROM idealista i
INNER JOIN agencias a 
    ON i.AGENCY_ID = a.AGENCY_ID
GROUP BY a.AGENCY_NAME
ORDER BY Precio_promedio DESC
LIMIT 1

"""
pd.read_sql(sSQL, con=conn)
```

### 5. Muestra todos los inmuebles y, si existe, el nombre de su agencia.

```python
sSQL = """

SELECT 
    i.ASSETID, 
    COALESCE(a.AGENCY_NAME, 'No tiene nombre de agencia') AS Nombre_Agencia
FROM idealista i
LEFT JOIN agencias a 
    ON i.AGENCY_ID = a.AGENCY_ID;

"""


pd.read_sql(sSQL, con=conn)
```

### 6. ¿Cuántos inmuebles gestiona cada agencia? Ordénalo de mayor a menor.

```python
sSQL = """

SELECT a.AGENCY_NAME as Agencia, COUNT(i.ASSETID) AS Total_inmuebles
FROM idealista i
INNER JOIN agencias a 
    ON i.AGENCY_ID = a.AGENCY_ID
GROUP BY i.AGENCY_ID, a.AGENCY_NAME
ORDER BY Total_inmuebles DESC

"""
pd.read_sql(sSQL, con=conn)
```
